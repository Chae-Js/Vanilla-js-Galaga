# 데이터 타입
- 데이터 타입은 값의 종류를 말한다. 
- 자바스크립트의 모든 값은 데이터 타입을 갖는다.
- 자바스크립트에는 총 7개의 타입이 있으며 이는 원시 타입과 객체 타입으로 나뉜다.(원시타입 6 + 객체 1 = 7개)   
\- 원시 타입: 숫자 타입, 문자열 타입, 불리언 타입. undefined 타입, null 타입, 심벌 타입  
\- 객체 타입: 객체, 함수, 배열 등

## 숫자 타입
- C, 자바 등과 달리 자바스크립트에는 하나의 숫자 타입만 존재한다.  
\- 실수, 정수, 음의 정수 모두 같은 숫자 타입일 뿐이다.  
\- 같은 숫자의 표기에 소수점이 붙든 붙지 않든 같은 수이다.  
\- 2진수. 8진수, 16진수 등을 담는 자료형이 없기 때문에 같은 수를 각각 2, 8, 16진수로 표기해도 같은 수이다.  
    ex1) 10 === 10.0
- 자바스크립트의 숫자 타입은 모든 수를 실수로 처리한다.  
\- 정수로 표시되는 수끼리 나누어도 실수가 나올 수 있다.  
    ex2) console.log(3/2) // 1.5
- 숫자 타입은 3가지 특별한 값을 표현할 수 있다.  
\- Infinity(양의 무한대), -Infinity(음의 무한대), NaN(산술 연산 불가: not-a-number)  
    ex3) console.log(10/0) // Infinity  
    ex4) console.log(10/-0) // -Infinity  
    ex5) console.log(1*'String') // NaN  

## 문자열 타입
- 문자열 타입은 텍스트 데이터를 나타내는 데 사용한다. 
- 0개 이상의 16비트 유니코드 문자(UTF-16)의 집합으로 전세계 대부분 문자를 표현 가능하다.
- 문자열은 작은 따옴표(''), 큰 따옴표(""), 백틱(\`\`)으로 텍스트를 감싼다.  
\- 가장 일반적으로는 작은따옴표를 사용한다.  
    ex6) var string;  
         string = '문자열';  
         string = "문자열";  
         string = `문자열`;
\- 작은따옴표로 감싼 문자열 안에서 큰따옴표는 문자열로, 큰따옴표로 감싼 문자열 안에서 작은따옴표는 문자열로 인식된다. 백틱도 이와 같다.
- 문자열을 따옴표, 백틱으로 감싸는 이유는 키워드, 식별자 등의 토큰과 구별하기 위해서이다. 또한 따옴표 없이는 공백 등도 포함할 수 없다.
- 문자열은 C에서는 문자의 배열, 자바에서는 객체이지만 자바스크립트에서는 원시 타입이며 변경 불가능한 값이다.

## 템플릿 리터럴
- ES6부터 도입된 문자열 표기법으로 멀티라인 문자열, 표현식 삽입, 태그드 템플릿 등의 편리한 문자열 처리 기능을 제공한다.
- 템플릿 리터럴은 런타임에 일반 문자열로 변환되어 처리된다.
- 템플릿 리터럴은 백틱(\`\`)을 이용해 표현한다.
### 멀티라인 문자열
- 일반 문자열 내에서는 줄바꿈, 들여쓰기 등을 위해서는 이스케이프 시퀀스를 사용해야 한다. 그러나 템플릿 리터럴 내에서는 이스케이프 시퀀스 없이도 줄바꿈이 허용되며 공백도 그대로 적용 된다.  
### 표현식 삽입
- 문자열은 문자열 연산자 "+"를 사용해 연결 가능하다.  
\- +는 피연산자 중 하나 이상이 문자열인 경우 문자열 연산자로 동작하고 그 외에는 덧셈 연산자로 동작한다.  
    ex6) var first = "ba";  
         var second = "bo";  
         console.log("I'm " + first + second); // I'm babo  
- 템플릿 리터럴 내에서는 표현식 삽입을 통해 간단히 문자열을 삽입할 수 있다 이를 통해 문자열 연산자보다 가독성 좋고 간편하게 문자열을 조합가능하다.
- 표현식 삽입을 위해서는 "${}"으로 표현식을 감싸야한다. 이때 표현식의 평가 결과가 문자열이 아니더라도 문자열로 타입이 강제로 변환되어 삽입된다.  
    ex7) var first = "ba";  
         var second = "bo";  
         console.log(`I'm ${first}${second}`); // I'm babo
- 표현식 삽입은 반드시 템플릿 리터럴 내에서 사용해야 한다. 일반 문자열에서는 문자열의 일부로 취급 받는다.

## 불리언 타입
- 불리언 타입의 값은 논리적 참, 거짓을 나타내는 true, false가 있다.
- 주로 프로그림의 흐름을 제어하는 조건문에서 사용된다.

## undefined 타입
- undefined 타입의 값은 undefined 하나이다.
- var 키워드로 선언한 변수는 암묵적으로 undefined로 초기화 된다.(이를 통해 쓰레기 값이 들어있는 공간(빈공간)을 undefined로 초기화한다)
- 변수 참조 시 undefined가 반환된다면 이는 값이 할당되지 않았음을 알게 해준다.  
\- 자바스크립트 엔진이 변수를 초기화하는데 사용하는 undefined를 의도적으로 할당하는 것은 본래 취지와 어긋나며 혼란을 줄 수 있으므로 권장하지 않는다.

## null 타입
- null 타입의 값은 null 하나이다.
- 프로그래밍 언어에서 null은 변수에 값이 없다는 것을 의도적으로 명시(의도적 부재)할 때 사용한다.
- 변수에 null을 할당하는 것은 변수가 이전에 참조하던 값을 더 이상 참조하지 않겠다는 의미이다.  
\- 이는 이전에 할당되어 있던 값에 대한 참조를 명시적으로 제거하는 것을 의미하며 자바스크립트 엔진은 누구도 참조하지 않는 메모리 공간에 대해 가비지 콜렉션을 수행할 것이다.  
    ex8) var foo = "lee";  
         foo = null; // 이전 참조를 제거하며 "foo"는 더이상 lee를 참조하지 않는다.  
                     // null을 할당하는 것보다 변수의 스코프를 좁혀 변수를 빨리 소멸시키는 것이 더 낫다.
- 함수가 유효한 값을 반환할 수 없는 경우 명시적으로는 null을 반호나하기도 한다.  
\- HTML요소를 검색해 반환하는  document.querySelector 메서드는 조건에 부합하는 HTML 요소를 검색할 수 없는 경우에 에러 대신 null을 반환한다.

## 심벌 타입
- ES6에서 추가된 7번째 타입으로, 변경 불가능한 원시타입의 값이다.
- 심벌 값은 다른 값과 중복되지 않는 유일무이한 값이다. -> 주로 이르밍 충돌할 위험이 없는 객체의 유일한 프로퍼티 키를 만들기 위해 사용한다.
- 심벌 이외의 원시 값은 리터럴을 통해 생성하지만 심벌은 Symbol 함수를 호출해 생성한다.  
\- 이때 생성된 심벌 값은 외부에 노출되지 않으며, 다른 값과 절대 중복되지 않는 유일무이한 값이다.  
    ex9) var key_ = Symbol('key');  
         console.log(typeof key_); // symbol  
         
         var obj = {};  // 객체 생성
         obj[key_] = "value"; // 프로퍼티 키로 심벌 사용
         console.log(obj) // { [Symbol(key)]: 'value' }
         console.log(obj[key_]) // value

## 객체 타입
- 자바스크립트의 데이터 타입은 크게 원시와 객체 타입으로 분류된다. 
- 자바스크립트는 객체 기반 언어이며, 자바스크립트를 이루는 거의 모든 것이 객체이다.
- 앞의 6개 타입을 제외하고는 모든 데이터 값은 객체 타입이다.

## 데이터 타입의 필요성
### 데이터 타입에 의한 메모리 공간의 확보와 참조
- 값은 메모리에 저장하고 참조할 수 있어야 한다. 메모리에 값을 저장하려면 먼저 확보해야 할 메모리 공가의 크기를 결정해야 한다. 즉 몇 바이트의 메모리 공간을 사용해야 낭비와 손실 없이 값을 저장할 수 있는지 알아야 한다.
- 자바스크립트 엔진은 데이터 타입에 따라 정해진 크기의 메모리 공간을 확보한다. -> 변수에 할당되는 값의 데이터 타입에 따라 확보해야 할 메모리 공간의 크기가 결정된다.  
ex10) var score = 100;  
      위 코드에서 자바스크립트 엔진은 리터럴 100을 숫자 타입의 값으로 해석하고 숫자 타입의 값 100을 저장하기 위해 8바이트의 메모리 공간을 확보한다. 그리고 100을 2진수로 저장한다.    
      \- ECMAScript에서 문자열, 숫자 타입을 제외하고 데이터타입의 크기는 규정되지 않았다. 따라서 엔진 제조사에 따라 데이터 타입에 따라 확보되는 메모리 공간의 크기는 다를 수 있다.  
- 자바스크립트 엔진에서 값을 참조할때 데이터 타입에 따라 한번에 받아들여야 하는 메모리의 크기를 판단한다.  
\- 만약 식별자 score를 통해 숫자타입의 값 100이 저장된 메모리 공간의 주소를 찾아갈 때, 이때 우리는 값 100이 저장되어 있는 메모리 공간의 선두 메모리 셀의 주소를 찾아갈 수 있다. 이때 숫자 타입의 값은 8바이트, 즉 메모리 셀 8개이기 때문에 8개에 맞게 읽어 들이지 않으면 값이 손상된다. 데이터 타입을 통해 읽어들일 메모리 공간의 크기를 결정하고 그에 맞게 참조하는 것이다.

\- 컴파일러, 인터프리터는 심벌 테이블이라고 부르는 자료구조를 통해 식별자를 키로 바인딩된 값의 메모리 주소, 데이터 타입, 스코프 등을 관리한다.
### 데이터 타입에 의한 값의 해석
- 모든 값은 데이터 타입을 가지며, 메모리에 2진수(비트의 나열)로 저장된다. 메모리에 저장된 값은 데이터 타입에 따라 다르게 해석될 수 있다.  
    ex11) 메모리에 저장된 값 0100 0001을 숫자로 해석하면 65, 문자열로 해석하면 'A'다.
- 메모리에서 읽어 들인 2진수를 어떻게 해석할지 데이터 타입을 기준으로 결정한다.

## 동적 타이핑
### 정적 타입 언어
- "정적 타입 언어"는 변수를 선언할 때 변수에 할당할 수 있는 데이터 타입을 사전에 선언해야 한다.(명시적 타입 선언)
- 정적 타입 언어는 변수의 타입을 변경할 수 없으며, 변수에 선언한 타입에 맞는 값만 할당할 수 있다.
- 정적 타입 언어는 컴파일 시점에 "타입 체크"(선언한 데이터 타입에 맞는 값을 할당했는지 검사하는 처리)를 수행한다. 만일 타입 체크를 통과하지 못했다면 에러를 발생시키고 프로그램의 실행 자체를 막는다.   
\- 이를 통해 타입의 일관성을 강제함으로써 더욱 안정적인 코드의 구현을 통해 런타임에 발생하는 에러를 줄인다.
### 동적 타입 언어
- 자바스크립트는 변수 선언 시에 타입을 선언하지 않고 var, let, const키워드를 통해 변수를 선언한다. 어떤 데이터 타입의 값도 자유로이 할당 가능하다.
- 자바스크립트에서는 값을 할당하는 시점에 변수의 타입이 동적으로 결정되고 변수의 타입을 언제든지 자유롭게 변경할 수 있다.
- 자바스크립트에서 변수는 선언이 아닌 할당에 의해 타입이 결정(타입 추론)된다. 재할당에 의해 변수의 타입은 언제든 동적으로 변할 수 있다. 이런 특징을 "동적 타이핑"이라 하고 이러한 언어를 "동적 타입 언어"라고 한다.
### 동적 타입 언어와 변수
- 동적 타입 언어는 변수에 어떤 데이터 타입의 값이라도 자유롭게 할당 가능하다. 이러한 동적 타입 언어의 특징은 데이터 타입에 대해 편리하다는 것이다. 
- 동적 타입 언어에서 변수 값은 언제든지 변경될 수 있기 때문에 복잡한 프로그램에서 변화하는 변수 값을 추적하기 어려울 수 있다.
- 동적 타입 언어의 변수는 값에 변경에 따라 언제든지 타입이 변경 가능하므로, 동적 타입 언어의 변수는 값을 확인하기 전에 타입을 확실 할 수 없다.  
\- 자바스크립트는 개발자의 의도와 관계없이 자바스크립트 엔진에 의해 암묵적으로 타입이 변경될 수 있다.  
\- 동적 타입 언어는 유연성은 높지만 신뢰성은 떨어진다.
- 위와 같은 이유로 안정적인 프로그램을 만들기 위해 이전의 데이터 타입을 체크해야 하는데 이는 번거로우며 코드의 양도 증가시킨다. 따라서 변수를 사용할 때 주의할 사항은 다음과 같다.  
    1. 변수는 필요한 경우에 한해 제한적으로 사용한다.  
    \- 변수 값은 재할당에 의해 언제든 변경 가능하다. 이로 인해 동적 타입 언어인 자바스크립트에서는 타입을 잘못 예측해 오류가 발생할 수 있다. 변수의 개수가 많을수록 오류 발생 확률도 높아지며 변수의 무분별한 남발은 금물이며, 필요한 만큼만 최소한으로 유지하도록 해야 한다.  
    2. 변수의 유효 범위(스코프)는 최대한 좁게 만들어 변수의 부작용을 억제해야 한다.  
    \- 스코프가 넓을 수혹 변수로 인한 오류의 발생 확률이 높아진다.  
    3. 전역 변수의 사용은 최대한 자제한다.  
    \- 어디서든 참조, 변경 가능한 전역 변수는 의도치 않게 값이 변경될 확률이 높고 다른 코드에 영향을 줄 가능성도 높다. 따라서 전역 변수는 프로그램의 복잡성을 증가시키고 처리 흐름을 추적하기 어렵게 만들고 오류의 원인 특정을 어렵게 한다.  
    4. 변수보다는 상수를 통해 값의 변경을 억제한다.  
    5. 변수 이름은 변수의 목적, 의미를 파악할 수 있도록 네이밍한다.  
    \- 모든 식별자는 존재 이유를 파악할 수 있게 네이밍해야 한다. 식별자의 스코프가 넓을 수록 명확한 이름을 붙이자. 명확한 네이밍으 코드를 이해하기 쉽게 하고, 협업, 생산성 향상에 도움을 준다.
