# 표현식과 문

## 값
- 값은 식(표현식)이 평가되어 생성된 결과를 말한다.  
\- 평가란 식을 해석해서 값을 생성하거나 참조하는 것을 말한다.
- 모든 값은 데이터 타입을 가지며, 메모리에 2진수(비트)의 나열로 저장된다. 메모리에 저장된 값은 데이터 타입에 따라 다르게 해석 가능하다.
- 변수는 하나의 값을 저장하기 위해 확보한 메모리 공간 자체 또는 그 메모리 공간을 식별하기 위해 붙인 이름이므로 변수에 할당되는 것은 값이다.  
ex1) var sum = 10 + 20;  
\- 위 코드에서 sum이란 변수에 할당되는 것은 10 + 20이 아니라 10 + 20이 평가된 결과인 숫자 30이다. 따라서 10 + 20은 할당 이전에 평가되어 값을 생성해야 한다.

## 리터럴
- 리터럴은 사람이 이해할 수 있는 문자 또는 약속된 기호를 사용해 값을 생성하는 표기법을 말한다.  
ex2) 3  
\- ex2)의 3은 단순 숫자가 아니라 숫자 리터럴이다. 사람이 이해할 수 있는 숫자를 이용해 숫자 리터럴3을 코드에 기술하면 자바스크립트 엔진은 이를 평가해 숫자 값 3을 생성한다.
- 자바스크립트 엔진은 런타임에 리터럴을 평가해 값을 생성한다. 즉 리터럴은 값을 생성하기 위해 미리 약속한 표기법이다.  
\- 정수, 부동소수점, 2진수, 8진수, 16진수, 문자열, 불리언, null, undefined, 객체, 배열, 함수, 정규표현식 등도 리터럴이 있다.  
        * 아마 값 또는 표현식(=값)에 해당하는 것들을 대입하지 않고 '2', '바보'같이 적은 것들은 전부 리터럴인듯

## 표현식
- 표현식은 값으로 평가될 수 있는 문이다. -> 즉 표현식이 평가되면 새로운 값을 생성하거나 기존 값을 참조한다.  
\- 앞서 살펴본 리터럴은 값으로 평가된다. -> 리터럴도 표현식이다.  
var score = 100;  
    \- 위 코드에서 100은 리터럴로 자바스크립트 엔진에 의해 평가되어 100이란 값을 생성한다.  
var score = 50 + 50;  
    \- 위 코드에서 50 + 50은 리터럴과 연산자로 이뤄져 있다. 하지만 50 + 50도 평가되어 숫자 값 100을 생성하므로 표현식이다.  
    \- "var score = 100", "var score = 50 + 50"이 표현식이 아니다. "var score"부분은 선언문이다. 표현식은 " = 100", " = 50 + 50"이다.   
score;  
    \- 변수 식별자를 참조하면 변수 값으로 평가된다. 식별자 참조는 값을 생성하지는 않지만 값으로 평가되므로 표현식이다.  
- 표현식은 리터럴, 식별자, 연산자, 함수 호출 등의 조합으로 이뤄질 수 있다.  
    ex3) 리터럴 표현식
        10, 'hello'
    ex4) 식별자 표현식(이미 선언되었다 가정)
        sum, person.name, arr[1]
    ex5) 연산자 표현식
        10 + 20, sum
    ex6) 함수/ 메서드 호출 표현식(이미 선언되었따 가정)
        square();
        person.getName();
- 표현식은 표현식을 평가된 값은 동등한 관계이다. -> 즉 표현식은 값처럼 사용 가능하다. 문법적으로 값이 위치할 수 있는 자리에는 표현식도 위치할 수 있다는 것을 의미한다.  
    ex7) var x = 1 + 2;
- 표현식은 또한 다른 표현식의 일부가 되어 새로운 값을 만들어 낼 수 있다.  
    ex8) var x = 1 + 3;
         x + 3; // --> 7

## 문
- 문이란 프로그램을 구성하는 기본 단위이자 최소 실행 단위이다. 문의 집합으로 이뤄진 것이 프로그램이며 문을 작성하고 순서에 맞게 나열하는 것이 프로그래밍이다.
- 문은 여러 토큰으로 구성된다. 토큰이란 문법적인 의미를 가지면 문법적으로 더 이상 나눌 수 없는 코드의 기본 요소를 의미한다.  
    ex8) var sum = 1 + 2;  
    위 코드에서 문은 "var sum = 1 + 2;"이고 토큰은 "var", "sum", "=", "1", "+", "2", ";"이다.
- 문을 명령문이라고도 부른다. 즉 문은 컴퓨터에 내리는 명령이다. 문이 실행되면 명령이 실행되고 무슨 일인가가 일어나게 된다.
- 문은 선언문, 할당문, 조건문, 반복문 등으로 구분 가능하다.  
    ex9) 선언문
        var x;
        function foo () {}  
    ex10) 할당문
        x = 5;
    ex11) 조건문
        if(x > 1){ console.log(x); }
    ex12) 반복문
        for (var i = 0; i < 2; i++) { console.log(i); }

## 세미콜론과 세미콜론 자동 삽입 기능
- 세미콜론(;)은 문의 종료를 나타낸다. --> 자바스크립트 엔진은 세미콜론으로 문이 종료한 위치를 파악하고 순차적으로 하나씩 문을 실행한다.
- 문을 끝낼때에는 세미콜론을 붙여야 하나, 0개 이상의 문을 중괄호로 묶은 코드 블록({...}) 뒤에는 세미콜론을 붙이지 않는다.  
\- 반복문, 조건문 등에 붙는 코드 블록({...})은 문의 종료를 의미하는 자체 종결성을 지닌다.
- 자바스크립트 엔진은 소스코드를 해석할 때 문의 끝이라 예측되는 지점에 세미콜론을 자동으로 붙여주는 "세미콜론 자동 삽입 기능"(ASI)을 가지고 있다. 따라서 세미콜론을 붙이는 것은 옵션이다.  
\- 세미콜론 자동 삽입 기능이 개발자의 예측과 일치하지 않는 경우가 있어 세미콜론은 필수적으로 붙여주는 것이 좋다.  
    ex13)   
    function foo () {
        return
            {}
       }
     위 코드에서 함수 내부 문에서 개발자는 " return {};"를 예상하고 작성했지만 세미콜론 자동 삽입 기능은 "return; {};"와 같이 동작한다.

## 표현식인 문과 표현식이 아닌 문
- 표현식은 문의 일부일 수도, 그 자체로 문일 수도 있다. 
- 표현식인 문과 표현식이 아닌 문을 구별하는 가장 간단하고 명료한 방법은 변수에 할당해 보는 것이다.  
\- 할당문은 자바스크립트에서 할당하는 값으로 평가된다. 따라서 할당문은 표현식이다.  
    ex14) x = 100;  
    위 코드에서 할당문 "x = 100"은 100으로 평가된다. 따라서   
    var foo = x = 100;  
    에서 foo에는 100이 할당된다.    
\- 선언문은 대표적인 표현식이 아닌 문이다.  
\- 크롬 개발자도구, node.js 등에서 표현식이 아닌 문을 실행하면 언제나 undefined를 출력한다. 표현식인 문을 실행하면 평가된 값을 출력한다. 이를 완료 값이라 하며 이는 표현식의 평가 결과가 아니다. 따라서 변수에 할당도, 참조도 불가능하다.(var x = 100; 처럼 표현식이 아닌 문과 표현식인 문이 섞여 있으면 이때는 undefined를 반환한다.)
